```python
import numpy as np
import math

sudoku_data = []
first_line = True
with open("datasets/sudoku.csv") as f:
    for line in f:
        if first_line == False :
            parts = line.split(",")
#             parts[1] = parts[1][:-1]
            length = len(parts[0])
            sudoku = np.zeros(length, int)
            i =0
            for num in parts[0]:
                sudoku[i] = int(num)
                i= i+1    
            sudoku_data.append(np.reshape(sudoku,(int(math.sqrt(length)),-1)))
        else:
            first_line = False
        
```

```python
counter = np.array()
for sudoku in sudoku_data:
    
    for letter in sudoku[0]:
        counter[letter] = counter[letter] +1

counter
```
